```python
import requests
import pandas as pd

url = "https://api.coingecko.com/api/v3/coins/bitcoin/market_chart"

params = {
    "vs_currency": "usd",
    "days": 365,        # âœ… FREE TIER LIMIT
    "interval": "daily"
}

response = requests.get(url, params=params)
print("Status Code:", response.status_code)

data = response.json()

# Validate response
if "prices" not in data:
    raise ValueError(f"Invalid response: {data}")

prices = data["prices"]
volumes = data["total_volumes"]

# Create DataFrame
df = pd.DataFrame(prices, columns=["timestamp", "price"])
df["volume"] = [v[1] for v in volumes]

# Convert timestamp
df["date"] = pd.to_datetime(df["timestamp"], unit="ms")

df = df[["date", "price", "volume"]]

print("Rows fetched:", df.shape[0])
df.head()
```

```python
# Convert timestamp to datetime
df["open_time"] = pd.to_datetime(df["open_time"], unit="ms")

# Convert price & volume columns to float
price_cols = ["open", "high", "low", "close", "volume"]
df[price_cols] = df[price_cols].astype(float)

# Keep only useful columns
df = df[["open_time", "open", "high", "low", "close", "volume"]]

# Rename column
df.rename(columns={"open_time": "date"}, inplace=True)

df.info()
```

```python
import os

raw_path = "/content/drive/MyDrive/btc_project/data/raw"
os.makedirs(raw_path, exist_ok=True)

file_path = f"{raw_path}/btc_raw_data.csv"
df.to_csv(file_path, index=False)

print("Saved successfully at:", file_path)
```
