```python
import pandas as pd

raw_file_path = "/content/drive/MyDrive/btc_project/data/raw/btc_raw_data.csv"

df = pd.read_csv(raw_file_path)
df.head()
```

```python
print("Shape:", df.shape)
df.info()
```

```python
df["date"] = pd.to_datetime(df["date"])

# Always sort time-series data
df = df.sort_values("date").reset_index(drop=True)

df.head()
```

```python
df.isnull().sum()
```

```python
df.rename(columns={"price": "close"}, inplace=True)
```

```python
df["daily_return"] = df["close"].pct_change()

df.head()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.describe()
```

```python
import os

processed_path = "/content/drive/MyDrive/btc_project/data/processed"
os.makedirs(processed_path, exist_ok=True)
```

```python
processed_file_path = f"{processed_path}/btc_processed_data.csv"

df.to_csv(processed_file_path, index=False)

print("Processed data saved at:")
print(processed_file_path)
print("Final shape:", df.shape)
```
